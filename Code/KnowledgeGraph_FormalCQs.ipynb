```python
from rdflib import *
```

```python
automaticallyAvatargraph = Graph()  #create a graph
```

```python
rdfuri = "http://www.w3.org/1999/02/22-rdf-syntax-ns#"
rdf = Namespace(rdfuri)
dbpuri ="http://dbpedia.org/page/"
dbp= Namespace(dbpuri)
wikiuri = "https://www.wikidata.org/wiki/"
wiki = Namespace(wikiuri)
viafuri = "https://viaf.org/viaf/"
viaf = Namespace(viafuri)
geonamesuri = "https://www.geonames.org/"
geonames = Namespace(geonamesuri)
owluri = "	http://www.w3.org/2002/07/owl#"
owl = Namespace(owluri)
```

```python
automaticallyAvatargraph.bind("av", avatar)
automaticallyAvatargraph.bind("rdf", rdf)
automaticallyAvatargraph.bind("dbp", dbp)
automaticallyAvatargraph.bind("wiki", wiki)
automaticallyAvatargraph.bind("viaf", viaf)
automaticallyAvatargraph.bind("geo", geonames)
automaticallyAvatargraph.bind("owl", owluri)
```

```python
for item in startingdata:
  if "av:" in item:
    itemuri = URIRef(item.replace("av:", namespaceavatar))
    
    for literal in startingdata[item]["literals"]:
      if type(startingdata[item]["literals"][literal]) == str or type(startingdata[item]["literals"][literal]) == int:
        lit = str(startingdata[item]["literals"][literal])
        automaticallyAvatargraph.add((itemuri, URIRef(literal.replace("av:", namespaceavatar)), Literal(lit)))
      else:
        for multiobject in startingdata[item]["literals"][literal]:
          lit = str(multiobject)
          automaticallyAvatargraph.add((itemuri, URIRef(literal.replace("av:", namespaceavatar)), Literal(lit)))
 
    for uri in startingdata[item]["uris"]:     
      if type(startingdata[item]["uris"][uri]) == str:
          uristring = startingdata[item]["uris"][uri]
          if "av:" in uristring:
            uristring = uristring.replace("av:", namespaceavatar)
          if "rdf:" in uristring:
              uristring = uristring.replace("rdf:", rdfuri)
          if "dbp:" in uristring:
              uristring = uristring.replace("dbp:", dbpuri)
          if "wiki:" in uristring:
              uristring = uristring.replace("wiki:", wikiuri)
          if "viaf:" in uristring:
              uristring = uristring.replace("viaf:", viafuri)
          if "geo:" in uristring:
              uristring = uristring.replace("geo:", geonamesuri) 
         
          if "av:" in uri:
              automaticallyAvatargraph.add((itemuri, URIRef(uri.replace("av:", namespaceavatar)), URIRef(uristring))) 
          if "rdf:" in uri:
              automaticallyAvatargraph.add((itemuri, URIRef(uri.replace("rdf:", rdfuri)), URIRef(uristring)))
          if "owl:" in uri:
              automaticallyAvatargraph.add((itemuri, URIRef(uri.replace("owl:", owluri)), URIRef(uristring)))         
      else:
          for multiobject in startingdata[item]["uris"][uri]:
            uristring = multiobject.replace("av:", namespaceavatar)
            if "av:" in uri:
              automaticallyAvatargraph.add((itemuri, URIRef(uri.replace("av:", namespaceavatar)), URIRef(uristring))) 
            if "rdf:" in uri:
              automaticallyAvatargraph.add((itemuri, URIRef(uri.replace("rdf:", rdfuri)), URIRef(uristring)))
```
